```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

## Load Model

```python
tokenizer= GPT2Tokenizer.from_pretrained('gpt2-large')
model=GPT2LMHeadModel.from_pretrained("gpt2-large",pad_token_id=tokenizer.eos_token_id)
```

```python
tokenizer.decode(tokenizer.eos_token_id)
```

## Tokenize Sentences

```python
sentence="I like Ice Cream"
input_ids=tokenizer.encode(sentence, return_tensors='pt')
```

```python
tokenizer.decode(input_ids[0][2])
```

## Generate and Decode Text

```python
output=model.generate(input_ids,max_length=100,num_beams=5,no_repeat_ngram_size=2,early_stopping=True)
```

```python
output
```

```python
tokenizer.decode(output[0],skip_special_tokens=True)
```

## Output Result

```python
text=tokenizer.decode(output[0],skip_special_tokens=True)
```

## User Input

```python
sen=input("Plz Enter the statement: ")
in_ids=tokenizer.encode(sen, return_tensors='pt')
output=model.generate(in_ids,min_length=100,max_length=400,num_beams=5,no_repeat_ngram_size=2,early_stopping=True)
print(tokenizer.decode(output[0],skip_special_tokens=True))
```
